# Running data

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
running_2018 = pd.read_csv('Data/2018-running.csv', index_col='날짜')
running_2019 = pd.read_csv('Data/2019-running.csv', index_col='날짜')
running_2020 = pd.read_csv('Data/2020-running.csv', index_col='날짜')
running_2021 = pd.read_csv('Data/2021-running.csv', index_col='날짜')
```

```python
dataframes = [running_2018, running_2019, running_2020, running_2021]
running_df = pd.concat(dataframes)
```

# Graph 그리기

```python
mpl.rc('font', family='AppleGothic')
```

```python
running_graph_2018 = running_2018['거리'].plot(kind='line', 
                                             title='2018 running distance', 
                                             figsize=(20, 4), 
                                             legend=False, 
                                             fontsize=8, 
                                             color='orange') # Orange, Running Level 2
running_graph_2018.set_xlabel('날짜', fontsize=12)
running_graph_2018.set_ylabel('거리', fontsize=12)
plt.show()
```

```python
# running_graph_2018.dropna(inplace=True)
running_2018.plot(
              kind='scatter',
              x='거리',
              y='칼로리',
              color='orange',
              alpha=0.3,
              title='2018 running 거리 및 칼로리 상관관계'
              )
plt.show()
```
