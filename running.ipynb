# Graph 그리기

```python
mpl.rc('font', family='AppleGothic')

running_graph_2017 = running_2017['거리'].plot(kind='line', 
                                             title='2017 running distance', 
                                             figsize=(20, 4), 
                                             legend=False, 
                                             fontsize=8, 
                                             color='yellow') # Yellow, Running Level 1 (LOW)
running_graph_2017.set_xlabel('날짜', fontsize=12)
running_graph_2017.set_ylabel('거리', fontsize=12)
plt.show()
```

```python
# running_graph_2018.dropna(inplace=True)
running_2017.plot(
              kind='scatter',
              x='거리',
              y='칼로리',
              color='orange',
              alpha=0.3,
              title='2018 running 거리 및 칼로리 상관관계'
              )
plt.show()
```
